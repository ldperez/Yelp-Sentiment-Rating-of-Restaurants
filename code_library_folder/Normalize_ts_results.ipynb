```python
# Define a function to normalize the columns of a dataframe to 0-1 range
def normalize_df(df):
    # Copy the original dataframe
    normalized_df = df.copy()
    
    # Iterate over the columns of the dataframe, except for the 'location' column
    for col in normalized_df.columns:
        if col != 'location':
            # Get the minimum and maximum values of the column
            col_min = normalized_df[col].min()
            col_max = normalized_df[col].max()
            
            # Replace 0 values with col_min to preserve them after normalization
            normalized_df[col] = normalized_df[col].replace(0, col_min)
            
            # Normalize the column using the formula
            normalized_df[col] = (normalized_df[col] - col_min) / (col_max - col_min)
    
    # Return the normalized dataframe
    return normalized_df

# Call the function to normalize the dataframe
normalized_df = normalize_df(df)

# Print the normalized dataframe
normalized_df
```
